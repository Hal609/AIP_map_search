```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(
    {"figure.figsize": (17, 7)},
    style="ticks",
    palette=sns.color_palette("Set2"),
    color_codes=True,
    font_scale=0.8
)

%config InlineBackend.figure_format = "retina"
```

```python
bfs = pd.read_csv("results_bfs.csv")
dfs = pd.read_csv("results_dfs.csv")
ucs = pd.read_csv("results_ucs.csv")
a_star = pd.read_csv("results_a_star.csv")
```

```python
sns.regplot(bfs, x="Distance", y="BFS_Time", order=2, ci=95, label="BFS", scatter_kws={"alpha": 0.3})
sns.regplot(dfs, x="Distance", y="DFS_Time", order=2, ci=95, label="DFS", scatter_kws={"alpha": 0.3})
sns.regplot(ucs, x="Distance", y="UCS_Time", order=2, ci=95, label="UCS", scatter_kws={"alpha": 0.3})
sns.regplot(a_star, x="Distance", y="A_Star_Time", order=2, ci=95, label="A*", scatter_kws={"alpha": 0.3})

# Customize the plot
plt.xlabel("Heuristic Estimate to Goal")
plt.ylabel("Time /ms")
plt.title("2nd Order Polynomial Regression")
plt.legend()
plt.show()
```
