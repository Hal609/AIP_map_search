```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(
    {"figure.figsize": (17, 7)},
    style="ticks",
    palette=sns.color_palette("Set2"),
    color_codes=True,
    font_scale=0.8
)

%config InlineBackend.figure_format = "retina"

old_data = pd.read_csv("results_a_star_merged.csv")
new_data = pd.read_csv("results_a_star.csv")
```

```python
bfs_old = pd.read_csv("results_bfs_merged.csv")
bfs_new = pd.read_csv("results_bfs.csv")
```

```python
joined = pd.concat([bfs_old, bfs_new], axis=0)
joined = joined.reset_index()
joined = joined.drop(columns=["index", "Unnamed: 0"])
joined
```

```python
joined.to_csv("results_bfs_merged.csv", index=False)
```

```python
bfs = pd.read_csv("results_bfs_merged.csv")
dfs = pd.read_csv("results_dfs.csv")
joined = pd.read_csv("results_a_star_merged.csv")
```

```python
sns.regplot(joined, x="Distance", y="A_Star_Time", order=2, ci=95, label="A*", scatter_kws={"alpha": 0.3})
sns.regplot(dfs, x="Distance", y="DFS_Time", order=2, ci=95, label="DFS", scatter_kws={"alpha": 0.3})
sns.regplot(bfs, x="Distance", y="BFS_Time", order=2, ci=95, label="BFS", scatter_kws={"alpha": 0.3})

# Customize the plot
plt.xlabel("Heuristic Estimate to Goal")
plt.ylabel("Time /ms")
plt.title("2nd Order Polynomial Regression")
plt.legend()
plt.show()
```
